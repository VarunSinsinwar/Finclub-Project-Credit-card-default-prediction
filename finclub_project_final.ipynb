**Feature Engineering**

1. Repayment Behavior Patterns

```python
pay_cols = ['pay_1', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']

# Max delay in any month
df['max_delay'] = df[pay_cols].max(axis=1)

# Count of months with delay (pay_m >= 1)
df['n_delinquent_months'] = df[pay_cols].apply(lambda row: (row >= 1).sum(), axis=1)

# Count of on-time full payments (pay_m == -1)
df['n_ontime_full'] = df[pay_cols].apply(lambda row: (row == -1).sum(), axis=1)

# Average delay over delinquent months
def avg_delay(row):
    delays = [x for x in row if x >= 1]
    return sum(delays) / len(delays) if delays else 0

df['avg_delay'] = df[pay_cols].apply(avg_delay, axis=1)

# Longest streak of consecutive overdue months
def longest_streak(row):
    streak = 0
    max_streak = 0
    for val in row:
        if val >= 1:
            streak += 1
            max_streak = max(max_streak, streak)
        else:
            streak = 0
    return max_streak

df['delinquency_streak'] = df[pay_cols].apply(longest_streak, axis=1)
```

2. Repayment Consistency & Discipline

```python
# Std deviation of payment status → inconsistency
df['pay_std'] = df[pay_cols].std(axis=1)

# Number of months with no usage (pay_m == -2)
df['n_no_usage'] = df[pay_cols].apply(lambda row: (row == -2).sum(), axis=1)

# Proportion of delinquent months
df['prop_delinquent'] = df['n_delinquent_months'] / 7
```

3. Utilization & Affordability

```python
# Credit utilization: avg bill / limit
df['credit_utilization'] = df['avg_bill_amt'] / df['LIMIT_BAL']

# Pay-to-limit ratio: total payment / limit
df['pay_to_limit'] = df[[f'pay_amt{i}' for i in range(1,7)]].sum(axis=1) / df['LIMIT_BAL']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, f1_score, ConfusionMatrixDisplay
```

```python
from sklearn.metrics import make_scorer, fbeta_score

# F2-score scorer
f2_scorer = make_scorer(fbeta_score, beta=2)
```

```python
from sklearn.model_selection import GridSearchCV

# -------------------- 1️⃣ Random Forest --------------------
rf = RandomForestClassifier(random_state=42)

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

grid_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf,
                       cv=5, scoring=f2_scorer, n_jobs=-1, verbose=2)

# -------------------- 2️⃣ XGBoost --------------------
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_xgb = GridSearchCV(estimator=xgb, param_grid=param_grid_xgb,
                        cv=5, scoring=f2_scorer, n_jobs=-1, verbose=2)

# -------------------- 3️⃣ LightGBM --------------------
lgb = LGBMClassifier(random_state=42)

param_grid_lgb = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, 15],
    'learning_rate': [0.01, 0.1, 0.2],
    'num_leaves': [31, 50, 100],
    'subsample': [0.8, 1.0]
}

grid_lgb = GridSearchCV(estimator=lgb, param_grid=param_grid_lgb,
                        cv=5, scoring=f2_scorer, n_jobs=-1, verbose=2)

# -------------------- Now Run the GridSearch --------------------

# Random Forest
print("Tuning Random Forest...")
grid_rf.fit(X_train_smote, y_train_smote)
print("Best RF params:", grid_rf.best_params_)

# XGBoost
print("\nTuning XGBoost...")
grid_xgb.fit(X_train_smote, y_train_smote)
print("Best XGB params:", grid_xgb.best_params_)

# LightGBM
print("\nTuning LightGBM...")
grid_lgb.fit(X_train_smote, y_train_smote)
print("Best LGB params:", grid_lgb.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression

# LOGISTIC REGRESSION
lr = LogisticRegression(max_iter=1000, solver='liblinear')

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

grid_lr = GridSearchCV(estimator=lr, param_grid=param_grid_lr,
                       scoring=f2_scorer, cv=5, n_jobs=-1, verbose=2)

# DECISION TREE
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

param_grid_dt = {
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

grid_dt = GridSearchCV(estimator=dt, param_grid=param_grid_dt,
                       scoring=f2_scorer, cv=5, n_jobs=-1, verbose=2)

# ADABOOST
from sklearn.ensemble import AdaBoostClassifier

adaboost = AdaBoostClassifier(random_state=42)

param_grid_ada = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}

grid_ada = GridSearchCV(estimator=adaboost, param_grid=param_grid_ada,
                        scoring=f2_scorer, cv=5, n_jobs=-1, verbose=2)


# Logistic Regression
grid_lr.fit(X_train_smote, y_train_smote)

# Decision Tree
grid_dt.fit(X_train_smote, y_train_smote)

# AdaBoost
grid_ada.fit(X_train_smote, y_train_smote)
```

```python

def evaluate(model, X_train, y_train, X_test, y_test, model_name=""):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None

    print(f"\n--- {model_name} ---")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\nTrain accuracy: ", accuracy_score(y_train, y_pred_train))
    print("Test accuracy: ", accuracy_score(y_test, y_pred) )
    if y_proba is not None:
        print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))

    print("F2 Score:", fbeta_score(y_test, y_pred, beta=2))
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_ if hasattr(model, 'classes_') else None)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()
```

```python
evaluate(grid_lr.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='Logistic Regression')
evaluate(grid_dt.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='Decision Tree')
evaluate(grid_ada.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='Adaboost')
evaluate(grid_rf.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='RandomForest')
evaluate(grid_xgb.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='XGBoost')
evaluate(grid_lgb.best_estimator_, X_train_smote, y_train_smote, X_test, y_test, model_name='LightGBM')
```

```python
from sklearn.metrics import fbeta_score, accuracy_score

def tune_threshold_with_accuracy(model, X_test, y_test, beta=2, model_name="Model"):
    y_proba = model.predict_proba(X_test)[:, 1]
    thresholds = np.arange(0.1, 0.9, 0.01)

    best_f2 = 0
    best_thresh = 0
    best_acc = 0

    for thresh in thresholds:
        y_pred = (y_proba >= thresh).astype(int)
        f2 = fbeta_score(y_test, y_pred, beta=beta)
        acc = accuracy_score(y_test, y_pred)

        if f2 > best_f2:
            best_f2 = f2
            best_thresh = thresh
            best_acc = acc

    print(f"✅ {model_name}")
    print(f"   Best F{beta}-Score: {best_f2:.4f}")
    print(f"   Accuracy at best threshold: {best_acc:.4f}")
    print(f"   Threshold: {best_thresh:.2f}\n")

    return best_thresh, best_f2, best_acc

```

```python
# Logistic Regression
thresh_lr, f2_lr, acc_lr = tune_threshold_with_accuracy(grid_lr.best_estimator_, X_test, y_test, model_name="Logistic Regression")

# Random Forest
thresh_rf, f2_rf, acc_rf = tune_threshold_with_accuracy(grid_rf.best_estimator_, X_test, y_test, model_name="Random Forest")

# XGBoost
thresh_xgb, f2_xgb, acc_xgb = tune_threshold_with_accuracy(grid_xgb.best_estimator_, X_test, y_test, model_name="XGBoost")

# LightGBM
thresh_lgb, f2_lgb, acc_lgb = tune_threshold_with_accuracy(grid_lgb.best_estimator_, X_test, y_test, model_name="LightGBM")
```
